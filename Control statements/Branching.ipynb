## Branching Statements
- These Statements are used to execute or skip of block of code based on the condition.

- The if statement is used to execute a block of code if a specified condition is true.


- if
- if else
- if elif
- nested if
- match case



```python
n=10
if n>1:
    print("Iron Man")
```

```python
# To check whether the integer is even or odd

n = int(input('n: '))
if n % 2==0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
n= int(input('n:  '))
if n==1:
    print("if")
elif n==2:
    print("elif")
elif n==3:
    print("nested if")
else:
    print("else")
```

```python
# Nested if

n= int(input("n: "))
if n % 3==0:
    if n % 7==0:
            print("number is divisible by both 3 and 7")
    else:
        print("number is divisible by 3, Not 7")
else:
    print("No. is not divisible by 3 and 7")
```

```python
# match Case
'''
match var:
case 1:
    #statement
Case 2:
    #statement
case _:
    # statement
'''
```

```python
num = int(input("Enter the Number"))
match num:
    case 1:
        print("Iron Man")
    case 2:
        print("Spider Man")
    case 3:
        print("Flash Man")
    case _:
        print("Buntyyyy")
```
