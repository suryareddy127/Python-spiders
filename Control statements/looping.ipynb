## Looping Statements
- These Statements are used to execute a block of code repeatedly until a certain condition is met.

- for loop
    - using range() function
    - using Iterables

- while loop

```python
# print the number from 1 to 10
 
for i in range(1,11,1):
    print(i)
```

```python
# generate the number from 1 to n

n= int(input())
for i in range(1,n+1):
    print(i,end=" ")
```

```python
#looping over list
l=[1,2,3,7,9,4,0,1]
for i in l:
    print(i)
```

```python
for i in range(len(l)):
    print(l[i])
```

```python
# create a list of natural numbers from 1 to 5
l=[]
for i in range (1,6):
    l.append(i)
print(l)
```

```python
# create a list of squares of even no's from 1 to 10
l=[]
for i in range(1,11):
    if i%2==0:
        l.append(i**2)
print(l)
```

```python
# create a list of squares with exisisting iterable
l=[1,2,3,7,9,4,0,1]
l1=[]
for i in l:
    if i%2==0:
        l1.append(i**2)
print(l1)
```
