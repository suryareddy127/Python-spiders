## Relational Operators

```python
a=7
b=3

print(a<b)  #less than
print(a>b) #greater than
print(a==b) #equalto
print(a<=b) #less than equalto
print(a>=b) #greater than equalto
print(a!=b) #not equalto
```

## Logical Operator

```python
a = 10
b = 20
c = 10
print(a==b and a==c)  # Both true then return True
print(a==b or a==c)    # any one of them true then return True
print( not a==c)    # reverse true = false
```

## Assignment Operator

```python
a=10
print(a)

a+=5    
print(a,"\n")  # a=15

a-=5
print(a,"\n")   #a=10

a*=5
print(a,"\n")   #a=50

a/=5
print(a,"\n")   #a= 10.0

a//=5
print(a,"\n")   #a= 2.0

a%=5
print(a,"\n")   #a= 2.0

a**=5
print(a,"\n")   #a= 32.0
```

## Membership Operator

1. in
    - It returns true if the element is a member otherwise it return False
2. not in
    - Not In returns true if the element is not a member otherwise it return false
- it return the bool result

```python
name="membership"
print('s' in name)
print('a' in name)
print('b' in name)
```

```python
name="membership"
print('s' not in name)
print('a' not in name)
print('b' not in name)
```
