### Strings Data type

- it is a data type which holds sequence of  characters within **('  ' , "  " , '''    ''')**

- it hold the characters such as cased and uncased
1. cased characters are uppercase & lower case alphabets
2. uncased characters are digits (0 to 9) and special symbols

```python
s= 'strings'
print(s,type(s),len(s))
```

```python
s1= "string's"
print(s1,type(s1),len(s1))
```

```python
s2= ''' String's "Data type" '''
print(s2,type(s2),len(s2))
```

- Strings are immutable in nature, It means cannot br modified
- strings are indexed collections ,it supports slicing

```python
s1= "pyspiders_btm"
print(s1[4])

# print(s1[13])   # IndexError: string index out of range

# s1[4]='r'
# print(s1)   # TypeError: 'str' object does not support item assignment
```

```python
s1= "pyspiders_btm"
print(s1[2:])

print(s1[:8])

print(s1[4:10])

print(s1[5:-3])

print(s1[-11:-4])

print(s1[5:100])

print(s1[-2:-9])  # no output because print statement will works from left to right,not from right to left
```

```python
s1= "pyspiders"
s2=" BTM"

print(s1 + s2) #concatenation
```

```python
print("dhoni" + 7)   #TypeError: can only concatenate str (not "int") to str
```

```python
print("dhoni" * "CSK")  # TypeError: can't multiply sequence by non-int of type 'str'
```

```python
print("dhoni " * 7)
```

ASCII values of Sting

'a' = 97        'A' = 65        '0' = 48
'z' = 122       'Z' = 90        '9' = 57

- ord(): it accept the string and returns the its respective ASCII value

```python
print(ord('a'))

print(ord("rahul"))    #TypeError: ord() expected a character, but string of length 5 found
```

- chr(integer): it accept the integer and converts into itsrespective ASCII Character

```python
print(chr(7))
print(chr(100))

print(chr('e'))   #TypeError: 'str' object cannot be interpreted as an integer
```
