## Tuple data type
- it is a collection datatype which stores multiple elements which are seperated by comma and enclosed with in ()
- var=(val1,val2,......,valn)
- for single element -> var= (val,)

```python
t=(1,2,3,4,5,7)
print(t,type(t))
```

```python
t=1,2,3,4,5,7
print(t,type(t))
```

```python
t=(1,2,3,7,"py",4,0,1)
print(t)
```

```python
#index
print(t[2])
print(t[-1])
```

```python
#slicing
print(t[1:3])
```

- It is a immutable datatype,
here no modifications on the elements of tuple datatype 

```python
t[1]= 70
print(t)
```

```python
print((1,2)+(2,3))

print((1,7)*4)
```

in built methods
1. count
2. index

```python
print(t)
print(t.count(1))

# print(t.count())  TypeError: tuple.count() takes exactly one argument (0 given)
```

```python
# tuple.index(value:[stop]:[start])
# it returns the index of first value

print(t.index(1))

print(t.index(1,3))

# print(t.index(8)) ValueError: tuple.index(x): x not in tuple
```
