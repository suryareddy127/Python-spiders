### List data type

- It is a collection datatype, which stores in a single variable enclosed within []
- It seperated by comma stored in a single variable enclosed within []

- Syntax- var[val1,val2,....,valn]


```python
l = [10,'py',70,80]
print(l,type(l))
```

```python
l1 = [10,False,70,80]
print(l1,type(l1))
```

```python
# len()

print(len(l))
```

```python
#index
print(l[1])
print(l[-2])
```

```python
#concatenation

print([1,2]+[3,4])

print([1,2]+[3,4]+[3,4])
```

```python
#multiplication
# - Sequnece*n

print([1,2]*2)
```

### In Built methods


1. Append
- It is a method which add a single element in the last position of a existing list
-  list.append(val)->None

```python
l=[10,90,50,60]
print(l)
l.append(100)
print(l)

l.append([1,7])
print(l)
```

```python
l.append(100,200)
l.append()  #TypeError: list.append() takes exactly one argument
#print(l)  #TypeError: list.append() takes exactly one argument (2 given)
```

2. Extend
- appends all elements from another list to the end of the current list.
- list.extend(iterable(list/tuple/str))->None

```python
l=[20,90,50,60]
l.extend([7,8,9])
print(l,"\n")
l.extend("py")
print(l)
l.extend([23,"py"])
l
```

3. insert
- Adds an element at a specific position (Before the index)
- list.insert(index,value)->None

```python
l=[10,30,"py",70]
l.insert(1,"python")
print(l)
l.insert(-2,'sql')
print(l)
```

```python
l.insert(1)  #
print(l)        #TypeError: insert expected 2 arguments, got 1
l.insert(1,2,3)
print(l)        #TypeError: insert expected 2 arguments, got 3
```

4. pop
- removes and returns the item at a given index
- list.pop(index)-> any
    - if not given the index value it will remove the last element

```python
l=[10,30,"py",70]
p=l.pop(2)
print(p,l)
print(l.pop())
```

```python
# l=[]
# print(l.pop(7))  IndexError: pop index out of range
# print(l.pop())  IndexError: pop from empty list
```

5. remove
- only removes the element from the list
- first occurence

```python
l=[10,20,30,40,70,90,10]
l.remove(10)
print(l)
# l.remove(8) ValueError: list.remove(x): x not in list

# l.remove(10,10)
# print(l)    TypeError: list.remove() takes exactly one argument (2 given)
```

```python
l=[10,20,30,40,70,[90,10]]
l.remove([90,10])
print(l)
```

6. copy

```python
l=[10,20,30,40,70,10,[90,10]]
print(l,id(l))
l1=l.copy()
print(l1,id(l1))
```

```python
l1.append(64)
print(l,id(l))
print(l1,id(l1))
```

```python
#clear
l1.clear()
print(l1)
```

```python
#count
print(l.count(10))
print(l.count(20))
print(l.count([90,10]))
# print(l.count(10,20))  TypeError: list.count() takes exactly one argument (2 given)
```

index
- list.index(value,start,stop) -> int(+ve)

```python
# index
l=[10,20,30,40,70,10]
print(l.index(10))
print(l.index(10,2))
# print(l.index([40]))  ValueError: [40] is not in list
```

```python
#reverse
l=[1,2,3,4,7,9]
print(l)
l.reverse()
print(l)
```

```python
# sort
l=[1,2,3,7,9,4,0,1]
l.sort()
print(l)
l.sort(reverse=True)
print(l)
```

### list comprehension

```python
print([i for i in range(1,11)])
```

```python
l=[i**2 for i in [1,2,3,7,9,4,0,1] if i%2==0]
print(l)
```

```python
#list comprehension using if-else

l=["even" if i%2==0 else "odd" for i in range(1,11)]
print(l)
```
