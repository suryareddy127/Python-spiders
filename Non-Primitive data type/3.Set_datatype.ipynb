## Set Datatype
- it is a collection datatype which stores multiple values seperated by , and enclosed within { }
- var={val1,val2,val3,......,valn}

```python
s= {12,45,89,76 }
print(s,type(s),len(s))
```

```python
# it is an unordered datatype
print(s,type(s),len(s))
```

```python
print(s[0])     #TypeError: 'set' object is not subscriptable
```

- sets are unordered its not posssible to provide the index
- it doesnot supports slicing
- it doesn't supoorts duplicate elements
- it is mutable datatype
- sset concatenations and multiplication is not possible, because set is not considered as a sequence.

```python
s= {12,45,89,76,12}
print(s,len(s))
```

### IN built methods

```python
# 1. add
s= {12,45,89,76}
s.add(79)
s.add(13)
print(s)

# s.add(2,3)  #TypeError: set.add() takes exactly one argument (2 given)
```

```python
# 2.pop
s.pop()
print(s)
```

```python
s1=s.pop()
print(s1)
s2=s.pop()
print(s2)
s3=s.pop()
print(s3)

```

```python
# 3. remove

s= {12,45,'py',90}
s.remove(90)
print(s)
```

```python
# 4. Discard
# it doesnot raises error if val is not present
s= {12,45,'py',90}
s.discard(34)
s
```

```python
# 5. copy
c=s.copy()
print(c,id(c))
```

```python
print(s,id(s))
```

```python
s.add(70)
print(s,id(s))
```

```python
# 6.clear
s.clear()
s
```

```python
# 7. union
s1= {12,45,'py',90}
s2= {11,43,'py',76}
s=s1.union(s2)
s
s3=s1.union([334,67],{23,77})
print(s3)

s4=s2.union("dog")
print(s4)
```

```python
# 8. update

s1.update(s2)
print(s1)

s1.update(s2,{1,3,4})
s1
```

```python
# 9. intersection

res= s1.intersection(s2)
res
```

```python
# 10. intersection update
s1.intersection_update(s2)
s1
```

```python
# 11.difference
s1= {12,45,'py',90}
s2= {11,43,'py',76}
s1.difference(s2)
s1
```
