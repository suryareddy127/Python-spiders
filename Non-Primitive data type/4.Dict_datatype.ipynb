### Dict Datatype
- it is a collection datatype which stores multiple values using key value pair which a pair of { }.

- var{key1:val1,key2:val2,.....,keyn:valn}

```python
d = {1:2,2:4,4:16,'a':'ant'}
print(d,type(d),len(d))
```

```python
# it accesing value use key  -> var[key]

print(d[1])
```

- it supports duplicate values but not duplicate keys

- if a key is repeated then its last value will overwritten its first value

```python
d1 = {1:2,2:4}
d1[1]="one"
print(d1)
```

- Note: in dictionary keys should be only off immutable datatypes **(int,float,complex,str,bool,tuple)** but values can be both mutable and immutable. 

In built methods

- view object : it is live or dynamic object where the changes on a dict are reflected automatically
1. keys() - returns the viewobject of keys
2. values() - returns the viewobject of values
3. items()  - returns the viewobject of items

```python
print(d.keys(),"\n")

print(d.values(),"\n")

print(d.items())
```

```python
for key,value in d.items():
    print(f"{key} : {value}")
```

```python
# Create a sample dictionary
student = {
    'name': 'Alice',
    'age': 20,
    'grade': 'A',
    'marks': 95
}

print("Method 1: Using .items()")
print("="*40)
for key, value in student.items():
    print(f"Key: {key} -> Value: {value}")
```
