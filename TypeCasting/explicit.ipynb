### Type Casing
- Converting one type of data into another type of data is called Type casting.
1. **Implicit** Type Casting ( user will change the type of data)
2. **Explicit** Type Casting ( Python itself Changing the type )

```python
a= input("A: ")
b= input("B: ")
print(a,b,sep=" ",)
```

```python
c= a + b
print(c,type(c))
```

```python
c= int(a) + int(b)
print(c,type(c))
```

```python
# str type casting 

s = str(34)
i = int(i)
f = float(i)
c = complex(i)
b = bool(i)

print(s,type(s))
print(i,type(i))
print(f,type(f))
print(c,type(c))
print(b,type(b))
```

```python
# Integer type casting 

i = 34       #defaultly allocated the type int
f = float(i)
c = complex(i)
b = bool(i)
print(i,type(i))
print(f,type(f))
print(c,type(c))
print(b,type(b))
```

```python
#  float type casting 

#f = 34.0
s = str(f)
i = int(f)
c = complex(f)
b = bool(f)

print(s,type(s))
print(i,type(i))
print(f,type(f))
print(c,type(c))
# print(b,type(b)
```

```python
# complex type casting

c = 34 + 0j
s= str(c)
i= int(c)
f = int(c)
b = bool(c)

print(c,type(c))
print(s,type(s))
print(i,type(i))
print(f,type(f))
print(b,type(b))

```

```python
a= "10.55"
int(a)          # ValueError: invalid literal for int() with base 10: '10.55'
float(a)        # 10.55
complex(a)      # (10.55+0j)
print(a,type(a))
```

```python
s= "string"
int(s)          #ValueError: invalid literal for int() with base 10: 'string'
float(s)        #ValueError: could not convert string to float: 'string'
complex(s)      #ValueError: complex() arg is a malformed string
```
