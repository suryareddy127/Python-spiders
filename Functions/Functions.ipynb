```python
### syntax
def function_name(parameters):
    """Doc String"""
    # function body
    return expression
```

```python
num = 7
if num %2==0:
    print("even")
else:
    print("odd")
```

```python
def check_num(num):
    if num %2==0:
        print("the number is even")
    else:
        print("the number is odd")
```

```python
#Functions with multiple parameters
def num(a,b):
    return a+b
res = num(3,4)
print(res)
```

```python
#default parameters in functions
def hello(name):
    print(f"hello {name}, welcome to the 10 days of python")
    
hello("surya")
```

```python
#*args ---Positinal arguments
def test(*args):
    for arg in args:
        print(arg)
test(1,2)
```

```python
#*kwargs ---Keyword arguments
def test(**kwargs):
    for key, value in kwargs.items():
        print(key, value)
test(sunday=1,monday=2)
```

```python
def numbers(*args, **kwargs):
    if args and kwargs:
        for arg in args:
            print(arg)
        for key, value in kwargs.items():
            print(f"{key}: {value}")
        print("Positional and Keyword arguments")
    elif args:
        for arg in args:
            print(arg)
        print("Positional arguments")
    elif kwargs:
        for key, value in kwargs.items():
            print(f"{key}: {value}")
        print("Keyword arguments")
    else:
        print("No arguments provided")
```

```python
numbers(1, 2, 3)
```

```python
numbers(sunday=1, monday=2)
```

```python
numbers(1, 2, 3, sunday=1, monday=2)
```

## Example of a functions

```python
#temperature conversion function
def temp_convert(temp,unit):
    """Celsius to Faherenheit"""
    if unit=="c":
        return (temp*9/5)+32
    elif unit=="f":
        return (temp-32)*5/9    #Faherenheit to Celsius

print(temp_convert(37,"c"))  
print(temp_convert(98,"f")) 
```

```python
#passoword Checker 
def password_checker(password):
    if len(password)<7:
        return False
    has_upper = False
    has_lower = False   
    has_digit = False
    
    for char in password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_digit = True    
    return has_upper and has_lower and has_digit
print(password_checker("Password123"))  # True
print(password_checker("pass"))         # False
    
```

```python
#palindrome checker
def palindrome(s):
    s=s.lower().replace(" ","")
    return s==s[::-1]
print(palindrome("Madam"))  # True
print(palindrome("Hello"))  # False
```

```python
#Calculate the factorial of a number using recursion
def factr(n):
    if n==0 or n==1:
        return 1
    else:   
        return n*factr(n-1)
print(factr(7))  # 5040
```
