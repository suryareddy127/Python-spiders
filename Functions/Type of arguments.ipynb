```python
def place_order(main_dish, side="Fries", *toppings, **details):
    print(f"Main Dish: {main_dish}")
    print(f"Side: {side}")
    
    if toppings:
        print(f"Toppings: {', '.join(toppings)}")
    
    if details:
        print("Special Instructions:")
        for key, value in details.items():
            print(f"- {key}: {value}")
    # print("-" * 20)
```

```python
# place_order("Burger") 

place_order("Pizza", "Salad", "Mushroom", "Olives", "Cheese",delivery="Express", orderno=5)

# place_order(delivery="Express", orderno=5)
```
